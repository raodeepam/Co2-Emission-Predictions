```python
sel_cols = ['country','year','co2']
df_req = df.select(sel_cols)
```

```python
df_req.show()
```

```python
pd_df_req = pd_df[['country','year','co2']]
pd_df_req
```

```python
import pandas as pd
pd_df_req['date'] = pd.to_datetime(pd_df_req['year'].astype(str) + '-01-01')
```

```python
pd_df_req.drop('year',axis=1,inplace=True)
```

```python
pd_df_req
```

```python
df_date_last = pd_df_req.loc[ (pd_df_req['date'] > '1970-01-01'), :]
```

```python
df_ind = df_date_last.loc[ (df_date_last['country'] == 'India') , :]
df_ind
```

```python
df_reg = df_ind
```

```python
df_reg['lag1'] = df_reg['co2'].shift(1)
df_reg['lag2'] = df_reg['co2'].shift(2)
df_reg['lag3'] = df_reg['co2'].shift(3)
df_reg['lag4'] = df_reg['co2'].shift(4)
df_reg['lag5'] = df_reg['co2'].shift(5)
```

```python
df_reg.dropna(inplace=True)
```

```python
df_reg_fin = df_reg.drop(['date','country'],axis=1)
```

```python
df_reg_fin
```

```python
sdf = spark.createDataFrame(df_reg_fin)
```

```python
sdf.printSchema()
```

```python
from pyspark.ml.feature import StandardScaler
from pyspark.ml.feature import VectorAssembler
```

```python
drop_col = ['co2']
input_data = sdf.select([column for column in sdf.columns if column not in drop_col])
```

```python
assembler = VectorAssembler(inputCols=input_data.columns, outputCol='features')
```

```python
from pyspark.ml import Pipeline

pipeline = Pipeline(stages=[assembler])

model = pipeline.fit(sdf)

sdf = model.transform(sdf)
```

```python
sdf.printSchema()
```

```python
selected_cols = ['features','co2']
sdf = sdf.select(selected_cols)
sdf.printSchema()
```

```python
training_data, test_data = sdf.randomSplit([0.7, 0.3], seed=7)
```

```python
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator


dt = DecisionTreeRegressor(featuresCol="features", labelCol="co2", maxDepth=5)
dt_model = dt.fit(training_data)

predictions = dt_model.transform(test_data)

# Evaluate the model's performance on the test data
evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="co2", metricName="rmse")
rmse = evaluator.evaluate(predictions)

print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
```
